```python
import warnings
warnings.filterwarnings("ignore")

import sys
sys.path.append('../')
sys.path.append('../src')
from src.main_worker import train_network

import plotly.express as px
import numpy as np
import pandas as pd
```

```python
out_data = train_network()
```

```python
epochs = np.arange(len(out_data['train_loss']))
df = {'epochs' : np.array([]), 'metric' : np.array([]), 'value': np.array([])}

for i, v in out_data.items():
    df['value'] = np.append(df['value'], v)
    df['epochs'] = np.append(df['epochs'], epochs)
    for _ in range(len(v)):
        df['metric'] = np.append(df['metric'], i)
```

```python
px.line(df, x="epochs", y='value', color='metric')
```
