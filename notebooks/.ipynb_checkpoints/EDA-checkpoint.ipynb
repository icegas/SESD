```python
def show_audios(df_path, r_esc, r_desed):
    x, sr = librosa.load(df_path["ESC"]+r_esc.filename)
    print("ESC: {}".format(sr) )
    x, sr = librosa.load(df_path["DESED"]+r_desed.filename)
    print("DESED: {}".format(sr) )

    a = ipd.Audio(df_path["ESC"]+r_esc.filename, rate=sr)
    display(a)
    a = ipd.Audio(df_path["DESED"]+r_desed.filename, rate=sr)
    display(a)
```

```python
#"ESC" : "DESED"
classes_map = { "pouring_water" : "Running_water", "vacuum_cleaner" : "Vacuum_cleaner" , 
                "dog" : "Dog", "Cat" : "cat" }
```

```python
df_path = {"DESED" : "..\data\DESED\eval\dataset\\audio\\eval\\public\\" ,
           "ESC" : "..\data\ESC-50-master\\audio\\"}

r_desed = df_desed.query('event_label=="Electric_shaver_toothbrush"').iloc[2]
r_esc = df_esc.query('category=="brushing_teeth"').iloc[2]
```

```python
show_audios(df_path, r_esc, r_desed)
```
